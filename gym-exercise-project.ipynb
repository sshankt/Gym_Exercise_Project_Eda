```python
sns.countplot(x = "Gender", data = df, color = "Red")
plt.title("Gender Count")
plt.show()
```

```python
df["Workout_Type"].value_counts()
```

```python
sns.countplot(x = "Workout_Type", data = df)
plt.title("Workout Count")
plt.show()
```

```python
plt.figure(figsize = (12,4))
sns.countplot(x = "Workout_Type" , hue = "Gender", data = df,  width = 0.8)
plt.title("Workout Type for each Gender")
plt.show()
```

```python
plt.figure(figsize = (12,4))
plt.bar(df["Workout_Type" ], df["Age"] , data = df, width = 0.7)
plt.title("Workout Type by Age")
plt.show()
```

```python
sns.barplot(x = "Workout_Type", y = "Calories_Burned", data = df)
plt.title("Workout Type influence on Calories Burned")
plt.show()
```

```python
sns.countplot(x="Workout_Type", hue="Gender", data=df)
plt.title("Workout Type by Gender")
plt.show()
```

```python
plt.figure(figsize =(10,4))
plt.bar(df["Gender"],df["Fat_Percentage"], color = "Orange")
plt.title("Fat Percentage by Gender")
```

```python
plt.bar(df["Gender"],df["Calories_Burned"])
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df["BMI"], bins=20, edgecolor="black", alpha=0.7)
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.boxplot(x="Workout_Type", y="BMI", data=df)
plt.title("BMI by Workout Type")
plt.xlabel("Workout Type")
plt.ylabel("BMI")
plt.show()
```

```python
plt.bar(df["Gender"],df["Height (m)"], width = 0.7, color = "green")
plt.xlabel("Gender")
plt.ylabel("Height")
plt.title("Weight by Gender")
plt.show()
```

```python
sns.boxplot(x='Experience_Level', y='Calories_Burned', data=df)
plt.title("Experince level by Calories Burned ")
plt.show()
```

```python
calories_by_gender = df.groupby("Gender")["Calories_Burned"].sum()
plt.pie(calories_by_gender, labels=calories_by_gender.index, autopct='%1.1f%%', startangle=90)
plt.title("Calories Burned by Gender")
plt.axis('equal')
plt.show()
```
